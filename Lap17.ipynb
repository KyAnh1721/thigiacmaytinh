```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Đọcảnh
image_path = 'C:/Users/Admin/Documents/thigiacmaytinh/cho07.png'
image = Image.open(image_path)

# Xámảnh
gray_image = image.convert('L')

# Chuyển đổi ảnh xám sang mạng
image_a = np.array(gray_image)

# Tính gradient theo hướng x
gradient_x = ndimage.sobel(image_a, axis=0, mode='constant')

#Displayảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Ảnhxốcgám')
plt.show()
```
