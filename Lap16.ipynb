```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
#Đườngdẫnđọcảnh
image_path = 'C:/Users/Admin/Documents/thigiacmaytinh/cho07.png'
image = Image.open(image_path)
```

```python
#Xảmánh
gray_image = image.convert('L')
```

```python
#Chuyểnxámảnhvềmảng
image_a = np.array(gray_image)
```

```python
#Bộ lọc Gaussian để làm mờ ảnh
blurred_image_a = ndimage.gaussian_filter(image_a, sigma=1, mode="constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma=4, mode="constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma=6, mode="constant")
```

```python
#Displayảnh
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xốc gám')

plt.subplot(2,2,2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('xámsigma1')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('xámsigma=3')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('xámsigma=5')
```
