```python
from PIL import Image
from imgtools import load_image
from IPython.display import display
```

```python
#Đườngdẫntớiảnh
my_path = 'C:/Users/Admin/Documents/thigiacmaytinh/dog.jpg'
```

```python
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 75, 175, 160)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```
